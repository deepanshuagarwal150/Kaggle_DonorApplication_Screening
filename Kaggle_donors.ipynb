```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
resources = pd.read_csv('resources.csv')
train.head()
```

```python
# Extracting info from project_submitted_datetime

train['project_submitted_datetime'] = pd.to_datetime(train['project_submitted_datetime'])
test['project_submitted_datetime'] = pd.to_datetime(test['project_submitted_datetime'])

train['year'] = train['project_submitted_datetime'].dt.year
train['month'] = train['project_submitted_datetime'].dt.month
train['weekday'] = train['project_submitted_datetime'].dt.weekday
train['date'] = train['project_submitted_datetime'].dt.day
train['hour'] = train['project_submitted_datetime'].dt.hour
train.drop('project_submitted_datetime',1, inplace= True)

test['year'] = test['project_submitted_datetime'].dt.year
test['month'] = test['project_submitted_datetime'].dt.month
test['weekday'] = test['project_submitted_datetime'].dt.weekday
test['date'] = test['project_submitted_datetime'].dt.day
test['hour'] = test['project_submitted_datetime'].dt.hour
test.drop('project_submitted_datetime',1, inplace= True)

train.head()
```

```python
# Extracting info from essay1, essay2, project_title, project_resource_summary

from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))

train['e1_length'] = train['project_essay_1'].apply(lambda x: len(str(x)))
train['e2_length'] = train['project_essay_2'].apply(lambda x: len(str(x)))
train['title_length'] = train['project_title'].apply(lambda x: len(str(x)))
train['summary_length'] = train['project_resource_summary'].apply(lambda x: len(str(x)))

train['text'] = train.apply(lambda row: ' '.join([str(row['project_essay_1']), str(row['project_essay_2'])]), axis=1)

# Extracting info from text
train['char_count'] = train['text'].apply(len)
train['word_count'] = train['text'].apply(lambda x: len(x.split()))
train['title_word_count'] = train['text'].apply(lambda x: len([word for word in x.split() if word.istitle()]))
train['upper_word_count'] = train['text'].apply(lambda x: len([word for word in x.split() if word.isupper()]))
train['stop_word_count']  = train['text'].apply(lambda x: len([word for word in x.split() if word.lower() in stopwords]))

# Test data

test['e1_length'] = test['project_essay_1'].apply(lambda x: len(str(x)))
test['e2_length'] = test['project_essay_2'].apply(lambda x: len(str(x)))
test['title_length'] = test['project_title'].apply(lambda x: len(str(x)))
test['summary_length'] = test['project_resource_summary'].apply(lambda x: len(str(x)))

test['text'] = test.apply(lambda row: ' '.join([str(row['project_essay_1']), str(row['project_essay_2'])]), axis=1)

# Extracting info from text
test['char_count'] = test['text'].apply(len)
test['word_count'] = test['text'].apply(lambda x: len(x.split()))
test['title_word_count'] = test['text'].apply(lambda x: len([word for word in x.split() if word.istitle()]))
test['upper_word_count'] = test['text'].apply(lambda x: len([word for word in x.split() if word.isupper()]))
test['stop_word_count']  = test['text'].apply(lambda x: len([word for word in x.split() if word.lower() in stopwords]))
```

```python
resources.head()
```

```python
# Prepairing new features from resources

resources['resources_total'] = resources['quantity']* resources['price']

res_sum = resources.groupby(['id'], as_index=False)[['resources_total']].sum()
res_mean = resources.groupby(['id'], as_index=False)[['resources_total']].mean()
res_mean = res_mean.rename(columns={'resources_total':'resources_total_mean'})
qty_sum = resources.groupby(['id'], as_index=False)[['quantity']].sum()
qty_sum = qty_sum.rename(columns={'quantity':'resources_quantity_sum'})
qty_count = resources.groupby(['id'], as_index=False)[['quantity']].count()
qty_count = qty_count.rename(columns={'quantity':'resources_quantity_count'})

train = pd.merge(train, res_sum, how='left', on = 'id').fillna(-1)
train = pd.merge(train, res_mean, how='left', on = 'id').fillna(-1)
train = pd.merge(train, qty_sum, how='left', on = 'id').fillna(-1)
train = pd.merge(train, qty_count, how='left', on = 'id').fillna(-1)

test = pd.merge(test, res_sum, how='left', on = 'id').fillna(-1)
test = pd.merge(test, res_mean, how='left', on = 'id').fillna(-1)
test = pd.merge(test, qty_sum, how='left', on = 'id').fillna(-1)
test = pd.merge(test, qty_count, how='left', on = 'id').fillna(-1)
```

```python
train.head()
```

```python
# gender mapping

gender = {"Mrs.": "Female", "Ms.":"Female", "Mr.":"Male", "Teacher":"Unknown", "Dr.":"Unknown", np.nan:"Unknown"  }
train['gender'] = train.teacher_prefix.map(gender)
test['gender'] = test.teacher_prefix.map(gender)
```

```python
# Converting categorical features into numerical features using Label Encoder

from sklearn import preprocessing
features = ['teacher_id', 'teacher_prefix', 'school_state', 'project_grade_category', 'project_subject_categories', 'project_subject_subcategories', 'gender']

df_combined = pd.concat([train[features], test[features]])

for feature in features:
    le = preprocessing.LabelEncoder()
    le = le.fit(df_combined[feature].astype(str))
    train[feature] = le.transform(train[feature].astype(str))
    test[feature] = le.transform(test[feature].astype(str))

```

```python
train.head()
```

```python
remove_cols = ['id', 'project_title', 'project_essay_1','project_essay_2','text','project_resource_summary']
train.drop(remove_cols, axis=1, inplace=True)
test.drop(remove_cols, axis=1, inplace=True)
train.head()
```

```python
remove_cols = ['project_essay_3','project_essay_4']
train.drop(remove_cols, axis=1, inplace=True)
test.drop(remove_cols, axis=1, inplace=True)
train.head()
```

```python
target = train.project_is_approved
train.drop('project_is_approved', 1, inplace= True)
```

```python
# Feature Scaling
# to bring all features in same range values
# this way one feature will not dominate another feature
# important step in Neural Nets and Deep Learning

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train = sc.fit_transform(train)
test = sc.fit_transform(test)
train
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=300, max_features='sqrt') # using 300 trees
model = model.fit(train,target)
```

```python
y_pred = model.predict(train)
y_pred
```

```python
# Checking Traning Accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(target, y_pred)
cm
```

```python
# Prediction on test data
test_prediction = model.predict(test)
test_prediction
```

```python
# Predict probabilities on test data as kaggle wants us to upload probability of category.
y_pred = model.predict_proba(test)
y_pred
```
